```python
import re
import numpy as np
import requests
import json
from pandas.io.json import json_normalize # Parse json as df
```

```python
def get_all_matches():
    
    # Filter to see which js script is refreshing in Chrome
    team_name_list = 'http://score.nowscore.com/data/bf.js?1576314086000'
    
    # Parse response and get soup object
    r = requests.get(team_name_list)
    soup = BeautifulSoup(r.text, 'lxml')
    
    # Regex match the pattern: A[i]=[content];
    pattern = re.compile(r'[A][[].+?[]][;]')

    # Get data by position
    match_ids = []
    match_time = []
    t1_cn = []
    t1_cn2 = []
    t1_eng = []
    t2_cn = []
    t2_cn2 = []
    t2_eng = []

    for i in re.findall(pattern, r.text):
        match_ids.append(i.split(']')[1].split('=')[1].split('[')[1].split(',')[0])
        match_time.append(i.split('[')[2].split(',')[10])
        t1_cn.append(i.split('[')[2].split(',')[4])
        t1_cn2.append(i.split('[')[2].split(',')[5])
        t1_eng.append(i.split('[')[2].split(',')[6])
        t2_cn.append(i.split('[')[2].split(',')[7])
        t2_cn2.append(i.split('[')[2].split(',')[8])
        t2_eng.append(i.split('[')[2].split(',')[9])
        
    # Check to see if any missing data

    assert(len(match_ids)==\
           len(match_time)==\
           len(t1_cn)==\
           len(t1_cn2)==\
           len(t1_eng)==\
           len(t2_cn)==\
           len(t2_cn2)==\
           len(t2_eng))
    
    # Construct df from lists

    data = list(zip(match_ids, match_time, t1_cn, t1_cn2, t1_eng, t2_eng, t2_cn2, t2_cn))

    df = pd.DataFrame(data, columns=['Match ID', 'Match Time', 
                                'Team 1 (CN)', 'Team 1 (CN-2)','Team 1 (Eng)',
                                'Team 2 (Eng)', 'Team 2 (CN-2)', 'Team 2 (CN)'])
    
    return df
```

## Unpack the json into dfs using keys

```python
def corner_kicks():
    #角球事件
    df = json_normalize(d['corner']['detail'])
    df['Team'] = df['ifHome'].apply(lambda x: 'Home' if x==1 else 'Away')
    return df.pivot_table(
        index='min',
        columns='Team'
    ).fillna('')
```

```python
def events():
    #詳細事件
    df = json_normalize(d['listItems'])
    
    def event_type(kind):
        if kind == 0:
            pass
        elif kind == 1:
            return 'Goal'
        elif kind == 2:
            pass
        elif kind == 3:
            return 'Yellow Card'
    
    df['Type'] = df['Kind'].apply(lambda x: events(x))
    
    return df
```

```python
def stats():
    #技術統計
    return json_normalize(d['listItemsTech'])
```

```python
def goals_conceded(option):
    #進失球概率 (choose between 30 or 50)
    
    if option == 30:
          
        df1 = json_normalize(d['JSQ']['JSQ_Home'])
        df2 = json_normalize(d['JSQ']['JSQ_Guest'])
        
    elif option == 50:
        
        df1 = json_normalize(d['JSQ_50']['JSQ_Home'])
        df2 = json_normalize(d['JSQ_50']['JSQ_Guest'])
    
    return df1.merge(df2, left_on='Time', right_on='Time')\
       .rename(columns={
        'JQ_x': '主隊進球',
        'SQ_x': '主隊失球',
        'JQ_y': '客隊進球',
        'SQ_y': '客隊失球'
    })
```

```python
goals_conceded(30)
```

```python
def technical_data():
    #技統數據
    df = json_normalize(d['JTL']).T
    
    df = df.reset_index().rename(columns={
        'index': 'Type',
        0: 'Value'
    })
    
    # Split the cols by delimiter _
    new_cols = df['Type'].str.split('_', n=1, expand=True)
    df['Type'] = new_cols[0]
    df['Type_Team'] = new_cols[1]

    df['Value'] = df['Value'].fillna(0).replace('', 0)
    
    return df.pivot_table(
        index='Type',
        columns='Type_Team'
    ).fillna('')
```

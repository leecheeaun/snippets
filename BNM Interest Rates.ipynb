```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd

%load_ext autoreload
%autoreload 2
```

### Get data from BNM website - Interest rates

```python
url = f"http://www.bnm.gov.my/index.php?csrf=763f5fd71de42b8080d3a6f690fd73be34361844&\
        ch=mmv&pg=stats_moneymarket&type=1&select=index.php%3Fch%3Dmmv%26pg%3Dstats_moneymarket%26type%3D1&\
        StartMth=1&StartYr=2000&EndMth=12&EndYr=2019"

agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}

page = requests.get(url, headers=agent)

soup = BeautifulSoup(page.content, 'lxml')
```

```python
def generate_df():

    df = []

    # Get header columns
    headers = [x.text.strip() for x in soup.find_all('th')]
    
    # Get data values
    for tr in soup.find_all('tr'):
        tds = tr.find_all('td')
        tds = df.append([x.text.strip() for x in tds])

    df = df[3:-1]
    
    # Concat the 2 dfs
    return pd.concat([pd.DataFrame(headers).T, pd.DataFrame(df)])
```

```python
df = generate_df()

# Promote 1st row and drop
df.columns = df.iloc[0]
df = df.iloc[1:,:].copy()

# Set datetime index
df.Date = pd.to_datetime(df.Date, format="%d %b %Y")
df.set_index('Date', inplace=True)

# Convert values to numeric
df.replace('-', '', inplace=True)
df = df.apply(pd.to_numeric)
```
